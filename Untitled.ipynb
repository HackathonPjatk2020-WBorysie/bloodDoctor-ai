# Averaging Names

```python
fruit_data = pd.read_excel("food_data.xlsx", index_col=0)
```

```python
fruit_data.head()
```

```python
fruit_data.tail()
```

```python
fruit_data["Shrt_Desc"].to_list()[0]
```

```python
merged_fruit_data = fruit_data.copy()
i = 0
for fruit in fruit_data.iloc[:,0]:
    stripped_name = fruit.split(",")[0]
    merged_fruit_data["Shrt_Desc"].iloc[i] = stripped_name
    i += 1
```

```python
merged_fruit_data.head()
```

```python
fruit_data.head()
```

# Merging Avg by Name

```python
merged_fruit_data["Shrt_Desc"].value_counts()
```

```python
merged_fruit_data.agg(["mean"])
```

```python
agg_data = merged_fruit_data.groupby("Shrt_Desc").agg("mean")
```

```python
agg_data.head()
```

# Filling Na's

```python
df = agg_data.copy()
```

```python
df = df.apply(lambda x: x.fillna(0) if x.dtype.kind in 'biufc' else x.fillna('.'))
```

```python
df.head()
```

# Calculating Std

```python
for col in df.columns:
    df[col].values[:] = agg_data[col].std() - agg_data[col]
```

```python
df.head()
```

# Second Nan's

```python
df = df.apply(lambda x: x.fillna(0) if x.dtype.kind in 'biufc' else x.fillna('.'))
```

```python
df.head()
```

# True Falsing Based on Threshold

```python
bin_data = df.copy()
```

```python
# bin_data.loc[bin_data["Water_(g)"] < 0, "Water_(g)"] = False
# bin_data.loc[bin_data["Water_(g)"] > 0, "Water_(g)"] = True
```

```python
bin_data.head()
```

```python
bin_data.columns
```

```python
bin_data["Water_(g)"].quantile(.4)
```

```python
for col in bin_data.columns:
    curr_val = bin_data[col]

    col_low = curr_val.quantile(.6)
    col_high = curr_val.quantile(.4)
    
    bin_data.loc[curr_val <= col_low, col] = False
    bin_data.loc[curr_val > col_high, col] = True
```
